##### **Eigenvalue Decomposition (Eigendecomposition)**
*Eigenvalue Decomposition* (ED) merupakan metode untuk memecah matriks persegi $m\times m$ menjadi bentuk perkalian antara eigenvektor dan eigenvalue. Dekomposisi ini umumnya dikenal dengan nama "Diagonalisasi Matriks."  
Dekomposisi eigenvalue dilakukan untuk membuat komputasi matriks menjadi lebih mudah, mengingat banyak aplikasi dari matriks seperti *machine learning* yang membutuhkan operasi matriks dengan dimensi yang sangat besar dan dilakukan secara berulang. Selain itu, dekomposisi matriks juga bisa diaplikasikan untuk menganalisis kestabilan suatu sistem dinamis. 

##### **Bentuk umum Eigendecomposition**
Eigendecomposition pada umumnya berbentuk seperti :
$$A = X \Lambda X^{-1}$$
$$A=
\begin{bmatrix}
\mid & \mid &  & \mid\\
x_1 & x_2 & \cdots & x_n\\
\mid & \mid &  & \mid
\end{bmatrix}
\begin{bmatrix}
\lambda_1 & 0 & \cdots & 0\\
0 & \lambda_2 & \cdots & 0\\
\vdots & \vdots & \ddots & \vdots\\
0 & 0 & \cdots & \lambda_n
\end{bmatrix}
\begin{bmatrix}
\mid & \mid &  & \mid\\
x_1 & x_2 & \cdots & x_n\\
\mid & \mid &  & \mid
\end{bmatrix}^{-1}$$


Dengan : 
- $A$ adalah matriks persegi
- $X$ adalah matriks persegi yang berisi eigenvektor
- $\Lambda$ adalah matriks diagonal yang berisi eigenvalue
- $X^{-1}$ adalah invers dari eigenvektor $X$

Apabila $A$ adalah matriks yang simetris, yakni ketika $A=A^{T}$. Maka, matriks $A$ memiliki beberapa properti yang menarik :  
- Eigenvalue-nya memiliki nilai real
- Eigenvektor-nya dapat dipilih untuk membentuk basis orthogonal
- Setiap matriks $A$ yang real dan simetris dapat didiagonalisasikan menjadi :
$$A = Q \Lambda Q^{T}$$
$$
A=
\begin{bmatrix}
\mid & \mid &  & \mid\\
q_1 & q_2 & \cdots & q_n\\
\mid & \mid &  & \mid
\end{bmatrix}
\begin{bmatrix}
\lambda_1 & 0 & \cdots & 0\\
0 & \lambda_2 & \cdots & 0\\
\vdots & \vdots & \ddots & \vdots\\
0 & 0 & \cdots & \lambda_n
\end{bmatrix}
\begin{bmatrix}
\mid & \mid &  & \mid\\
q_1 & q_2 & \cdots & q_n\\
\mid & \mid &  & \mid
\end{bmatrix}^{T}$$

Dengan :
- $Q$ adalah matriks orthogonal yang berisi eigenvektor 
- $\Lambda$ adalah matriks diagonal yang berisi eigenvalue
- $Q^{T}$ adalah transpose dari matriks $Q$

##### **Geometric Perspective of ED**
Untuk menjelaskan perspektif geometris dari eigendecomposition kita recall bentuk umum eigendecomposition dari $A$ dimana $A$ adalah matriks berukuran $m \times m$ :  
$$A = X \Lambda X^{-1}$$
Eigenvektor ($x$) dari transformasi ini adalah vektor nonzero yang tidak berubah jalurnya selama transformasi, tetapi hanya menanjang, memendek atau berbalik arah (atau tetap sama) yang ditentukan oleh faktor skala $\lambda$ sebagai akibat dari transformasi. Secara sederhana, eigenvektor ($x$) berperan sebagai arah scaling, dan $\lambda$ sebagai faktor scaling.
Peran dari setiap matriks akan lebih mudah dipahami ketika melakukan plotting yang akan di bahas di bagian berikutnya.

##### **Singular Value Decomposition**
Singular value decomposition (SVD) adalah metode dekomposisi matriks untuk memecah sebuah matriks persegi $m \times m$ maupun persegi panjang $m \times n$ menjadi 3 matriks , yaitu $U$, $\Sigma$, dan $V^T$ . Dimana $U$ adalah matriks orthogonal yang berisi eigenvektor dari martiks $AA^T$, $V^T$ adalah transpose dari matriks orthogonal yang berisi eigenvektor dari $A^TA$ sementara $\Sigma$ adalah matriks diagonal dengan $r$ elemen yang memiliki nilai sama dengan akar dari eigenvalue positif $AA^T$ atau $A^TA$. SVD biasanya digunakan untuk pengolahan citra seperti kompresi gambar, sistem rekomendasi , dan pengurangan noise datasets.

##### **Bentuk Umum Singular Value Decomposition**
Singular Value Decomposition secara umum berbentuk :
$$A = U \Sigma V^T$$
$$
\underbrace{A}_{m \times n}=
\underbrace{
    \begin{bmatrix}
    \text{ }& \text{ } & \text{ }\\
    \text{ } & U & \text{ }\\
    \text{ } & \text{ } &  \text{ }\\
    \end{bmatrix}
}_{m \times m}
\text{ }
\underbrace{
    \begin{bmatrix}
    \text{ }& \text{ } & \text{ }\\
    \text{ } & \Sigma & \text{ }\\
    \text{ }& \text{ } & \text{ }\\
    \end{bmatrix}
}_{m \times n}
\text{ }
\underbrace{
    \begin{bmatrix}
    \text{ }& \text{ } & \text{ }\\
    \text{ } & V & \text{ }\\
    \text{ }& \text{ } & \text{ }\\
    \end{bmatrix}^T
}_{n \times n}
$$
Dengan :
- $U$ adalah matriks orthogonal $m \times m$ yang berisi eigenvektor $AA^T$ (Left singular vector)
- $\Sigma$ adalah matriks diagonal $m \times n$ yang berisi singular value $(\sqrt{\Lambda})$
- $V$ adalah matriks orthogonal $n \times n$ yang berisi eigenvektor $A^TA$ (Right singular vector)


##### **Geometric Perspective of SVD**
Untuk menjelaskan perspektif geometris dari SVD, kita recall bentuk umum SVD dari $A$ dimana $A$ adalah matriks berukuran $m \times n$ : 
$$
\underbrace{A}_{m \times n}=
\underbrace{
    \begin{bmatrix}
    \text{ }& \text{ } & \text{ }\\
    \text{ } & U & \text{ }\\
    \text{ } & \text{ } &  \text{ }\\
    \end{bmatrix}
}_{m \times m}
\text{ }
\underbrace{
    \begin{bmatrix}
    \text{ }& \text{ } & \text{ }\\
    \text{ } & \Sigma & \text{ }\\
    \text{ }& \text{ } & \text{ }\\
    \end{bmatrix}
}_{m \times n}
\text{ }
\underbrace{
    \begin{bmatrix}
    \text{ }& \text{ } & \text{ }\\
    \text{ } & V & \text{ }\\
    \text{ }& \text{ } & \text{ }\\
    \end{bmatrix}^T
}_{n \times n}
$$

Ketika melakukan transformasi terhadap suatu titik/vektor/objek, langkah pertama adalah mengalikan titik/vektor/objek tersebut terhadap matriks $V^T$. Matriks ini berupa matriks orthogonal yang ditranspose, yang berarti berperan untuk melakukan rotasi ke arah basis standar di ruang $\R^n$.
Kemudian, hasil dari perkalian tersebut akan dikalikan dengan matriks $\Sigma$ yang merupakan matriks diagonal berisi singular value ($\sigma$). Matriks ini berperan sebagai faktor skala terhadap sumbu di ruang $\R^n$ (dalam kasus $\R^2$ maka sumbu-x dan sumbu-y). Dalam kasus matriks $A$ berukuran $m \times n$, matriks $\Sigma$ juga dapat berperan sebagai dimension eraser ketika $m < n$ atau dimension adder ketika $m > n$ yang akan membawa ke ruang $\R^m$.  
Matriks $U$ adalah matriks orthogonal, artinya matriks ini berperan untuk melakukan rotasi. Arah rotasi dari matriks $U$ yakni kearah left singular vector (matriks $U$) itu sendiri di ruang yang baru, yakni ruang $\R^m$. Penjelasan akan lebih mudah dipahami ketika melakukan plotting yang akan dilakukan di bagian berikutnya.

##### **Kasus 1  : Matriks persegi 2x2 nonsimetris**
Misalkan kita memiliki matriks $A$ berukuran $2 \times 2$ :
$$A =
\begin{bmatrix}
2 & 1 \\
0 & 3
\end{bmatrix}
$$
Kita akan lakukan eigendecomposition untuk matriks tersebut :

Pertama-tama, deklarasi fungsi terlebih dahulu untuk mempermudah programming :

```julia
using LinearAlgebra
using Plots

function plotting2D(title, a, b, colour, axislimit)
    plot(a, b, legend = false, aspect_ratio = :equal, title = title, color = colour, linewidth = 2,
    xlims = (-axislimit, axislimit), ylims = (-axislimit, axislimit), 
    xticks = -axislimit:1:axislimit, yticks = -axislimit:1:axislimit)
end

function plotvec(c, d)
    plot!([0,c[1]], [0,c[2]], arrow = :arrow, linewidth = 2, color = :blue)
    plot!([0,d[1]], [0,d[2]], arrow = :arrow, linewidth = 2, color = :red)
end

function plotting3D(title, a, b, c, colour, axislimit, azimuth, altitude)
    plot(a, b, c, legend = false, aspect_ratio = :equal, title = title, color = colour, linewidth = 2,
    xlims = (-axislimit, axislimit), ylims = (-axislimit, axislimit), zlims = (-axislimit,axislimit), 
    xticks = -axislimit:1:axislimit, yticks = -axislimit:1:axislimit, zticks = -axislimit:1:axislimit,
    camera = (azimuth, altitude))
end

function plotvec3D(c, d)
    plot!([0,c[1]], [0,c[2]], [0,c[3]], arrow = :arrow, linewidth = 2, color = :blue)
    plot!([0,d[1]], [0,d[2]], [0,d[3]], arrow = :arrow, linewidth = 2, color = :red)
end
```

Deklarasi matriks $A$ seperti berikut :

```julia
A = [2 1;0 3]
l, X = eigen(A)

L = Diagonal(l)
Xinv = inv(X)

x1 = X[:,1]
x2 = X[:,2]

display(X)
display(L)
display(Xinv)
```

Setelah melakukan eigendecomposition, dapat diketahui bahwa matriks $A$ dapat dipecah menjadi :
$$
A =
\underbrace{
    \begin{bmatrix}
    1 & 0.707107\\
    0 & 0.707107
    \end{bmatrix}
}_{X}
\underbrace{
    \begin{bmatrix}
    2 & 0\\
    0 & 3
    \end{bmatrix}
}_{\Lambda}
\underbrace{
    \begin{bmatrix}
    1 & -1\\
    0 & 1.41421
    \end{bmatrix}
}_{X^{-1}}
$$
*Note : Eigenvector ($X$) sudah dinormalkan.

Setelah Melakukan eigendecomposition, kita akan melakukan plotting kepada matriks tersebut dan melihat apa yang dilakukan matriks tersebut

```julia

t = range(0, 2π, length=100)
p = cos.(t)
q = sin.(t)

C = [p q]'

plotting2D("Unit circle", p, q, :viridis, 2)
plotvec(x1, x2)
```

Untuk mempermudah pengamatan, akan dilakukan plotting terhadap lingkaran serta eigenvektor $x_1 \text{ dan } x_2$ (vektor berwarna biru dan merah).

```julia
XinvC = Xinv*C
XinvC1 = XinvC[1,:]
XinvC2 = XinvC[2,:]

Xinvx1 = Xinv*x1
Xinvx2 = Xinv*x2
plotting2D("Xinv*C", XinvC1, XinvC2, :viridis, 2)
plotvec(Xinvx1, Xinvx2)
```

Setelah dikalikan dengan matriks $X^{-1}$, eigenvektor $x_1 \text{ dan } x_2$ dirotasi berorientasi sejajar dengan basis standar di $\R^2$

```julia
LXinvC = L*XinvC
LXinvC1 = LXinvC[1,:]
LXinvC2 = LXinvC[2,:]

LXinvx1 = L*Xinvx1
LXinvx2 = L*Xinvx2

plotting2D("L*Xinv*C", LXinvC1, LXinvC2, :viridis, 5)
plotvec(LXinvx1, LXinvx2)
```

Dapat diamati dari plot diatas, matriks diagonal $\Lambda$ akan melakukan scaling ke arah eigenvektor $x_1 \text{ dan } x_2$ dan mempertahankan orientasi eigenvektor $x_1 \text{ dan } x_2$ karena sejajar dengan basis standar.

```julia
XLXinvC = X*LXinvC
XLXinvC1 = XLXinvC[1,:]
XLXinvC2 = XLXinvC[2,:]

XLXinvx1 = X*LXinvx1
XLXinvx2 = X*LXinvx2
plotting2D("X*L*Xinv*C", XLXinvC1, XLXinvC2, :viridis, 4)
plotvec(XLXinvx1, XLXinvx2)
```

Eigenvektor $x_1 \text{ dan } x_2$ yang telah diskalakan, tetapi dengan orientasi yang tidak berubah, dikembalikan ke orientasi awalnya dengan mengalikannya dengan matriks $X$.

```julia
AC = A*C
AC1 = AC[1,:]
AC2 = AC[2,:]

Ax1 = A*x1
Ax2 = A*x2
plotting2D("A*C", AC1, AC2, :viridis, 4)
plotvec(Ax1, Ax2)
```

Keseluruhan proses diatas sama dengan mengalikan eigenvektor $x_1 \text{ dan } x_2$ maupun matriks lingkaran $C$ secara langsung.

##### **Kasus 2 : Matriks 2x2 Simetris**
Sekarang, akan dilakukan dekomposisi terhadap matriks simetris $B$ dengan ukuran $2 \times 2$ berikut :
$$
B = 
\begin{bmatrix} 
    5 & 3 \\
    3 & 9\
\end{bmatrix}
$$
Maka, eigendecompositionnya menjadi :

```julia
B = [5 3;3 9]

lm, Q = eigen(B)
Lm = Diagonal(lm)
QT = Q'
q1 = Q[1,:]
q2 = Q[2,:]

display(Q)
display(Lm)
display(QT)
```

Setelah dilakukan eigendecomposition, matriks $B$ dapat dipecah menjadi :
$$
B =
\underbrace{
    \begin{bmatrix}
    -0.881675 & 0.471858\\
    0.471858 & 0.881675
    \end{bmatrix}
}_{Q}
\underbrace{
    \begin{bmatrix}
    3.39445 & 0\\
    0 & 10.6056
    \end{bmatrix}
}_{\Lambda}
\underbrace{
    \begin{bmatrix}
    -0.881675 & 0.471858\\
    0.471858 & 0.881675
    \end{bmatrix}
}_{Q^{T}}
$$
Setelah Melakukan eigendecomposition, kita akan melakukan plotting kepada matriks tersebut dan melihat apa yang dilakukan matriks tersebut

```julia
plotting2D("Unit circle (C)", p, q, :orange, 2)
plotvec(q1, q2)
```

Dapat diamati untuk kasus matriks simetris $B$ memiliki eigenvektor $q_1 \text{ dan } q_2$ (berwarna merah dan biru) yang saling orthogonal dan dapat dibuat menjadi orhtonormal.

```julia
QTC = QT*C
QTC1 = QTC[1,:]
QTC2 = QTC[2,:]

QTq1 = QT*q1
QTq2 = QT*q2

plotting2D("Q^T*C", QTC1, QTC2, :orange, 2)
plotvec(QTq1, QTq2)
```

Setelah dikalikan dengan matriks $Q^{T}$, eigenvektor $q_1 \text{ dan } q_2$ dirotasi agar berorientasi sejajar dengan basis standar di $\R^2$

```julia
LmQTC = Lm*QTC
LmQTC1 = LmQTC[1,:]
LmQTC2 = LmQTC[2,:]

LmQTq1 = Lm*QTq1
LmQTq2 = Lm*QTq2

plotting2D("Lm*Q^T*C", LmQTC1, LmQTC2, :orange, 12)
plotvec(LmQTq1, LmQTq2)
```

Dapat diamati dari plot diatas, matriks diagonal $\Lambda$ akan melakukan scaling ke arah eigenvektor $q_1 \text{ dan } q_2$ dan mempertahankan orientasi eigenvektor $q_1 \text{ dan } q_2$ karena sejajar dengan basis standar.

```julia
QLmQTC = Q*LmQTC
QLmQTC1 = QLmQTC[1,:]
QLmQTC2 = QLmQTC[2,:]

QLmQTq1 = Q*LmQTq1
QLmQTq2 = Q*LmQTq2

plotting2D("Q*Lm*Q^T*C", QLmQTC1, QLmQTC2, :orange, 12)
plotvec(QLmQTq1, QLmQTq2)
```

Eigenvektor $q_1 \text{ dan } q_2$ yang telah diskalakan dengan orientasi yang tidak berubah, dikembalikan ke orientasi awalnya dengan mengalikannya dengan matriks $Q$.


##### **Kasus 3 : Matriks persegi panjang 3x2**
Misalkan kita memiliki matriks $C$ berukuran $3 \times 2$  seperti berikut :
$$
C=
\begin{bmatrix}
5 & 8\\
1 & 1\\
6 & 7
\end{bmatrix}
$$
Lalu kita akan melakukan SVD pada matriks tersebut seperti langkah di bawah ini :

```julia
C = [5 8; 1 1; 6 7]
U, s, V = svd(C, full=true)
S = [Diagonal(s); 0 0]

VT = V'
Cir = [p q]'

vt1 = VT[1,:]
vt2 = VT[2,:]

display(U)
display(S)
display(VT)
```

Setelah dilakukan dekomposisi, matriks $C$ bisa dipecah menjadi :
$$
C=
\underbrace{
    \begin{bmatrix}
    -0.711183 & 0.699022 & -0.0747435\\
    -0.105682 & -0.211417 & -0.971666\\
    -0.695018 & -0.683133 &  0.224231
    \end{bmatrix}
}_{U} \text{  }
\underbrace{
    \begin{bmatrix}
    13.2279 & 0.0\\
    0.0  &   1.01143\\
    0.0  &   0.0
    \end{bmatrix}
}_{\Sigma} \text{  }
\underbrace{
    \begin{bmatrix}
    -0.59206 &  -0.805894\\
    -0.805894 & 0.59206
    \end{bmatrix}
}_{V^T}
$$

Selanjutnya, kita akan melakukan plotting untuk mengetahui apa yang dilakukan oleh matriks $C$ tersebut :

```julia
plotting2D("Unit circle (Cir)", p, q, :green, 2)
plotvec(vt1, vt2)
```

Pertama-tama, kita mulai dengan unit circle $Cir$ dan vektor yang diamati adalah $v^T_1 \text{ dan } v^T_2$ (berwarna merah dan biru). 

```julia
VTCir = VT*Cir
VTCir1 = VTCir[1,:]
VTCir2 = VTCir[2,:]

VTvt1 = VT*vt1
VTvt2 = VT*vt2

plotting2D("VT*Cir", VTCir1, VTCir2, :green, 2)
plotvec(VTvt1, VTvt2)
```

Setelah dikalikan dengan matriks $V^{T}$, vektor $v^T_1 \text{ dan } v^T_2$ dirotasi berorientasi sejajar dengan basis standar di $\R^2$.

```julia
SVTCir = S*VTCir
SVTCir1 = SVTCir[1,:]
SVTCir2 = SVTCir[2,:]
SVTCir3 = SVTCir[3,:]

SVTvt1 = S*VTvt1
SVTvt2 = S*VTvt2

plotting3D("S*VT*Cir", SVTCir1, SVTCir2, SVTCir3, :green, 14, 70, 30)
plotvec3D(SVTvt1, SVTvt2)
```

Setelah mengalikan dengan matriks diagonal $\Sigma$, selain mengalami scaling, vektor juga akan berpindah ruang ke $\R^3$ karena pada matriks diagonal $\Sigma$ terdapat baris bernilai nol (baris ketiga) yang berperan sebagai dimension adder.

```julia
USVTCir = U*SVTCir
USVTCir1 = USVTCir[1,:]
USVTCir2 = USVTCir[2,:]
USVTCir3 = USVTCir[3,:]

USVTvt1 = U*SVTvt1
USVTvt2 = U*SVTvt2

plotting3D("U*S*VT*Cir", USVTCir1, USVTCir2, USVTCir3, :green, 10, 70, 30)
plotvec3D(USVTvt1, USVTvt2)
```

Setelah dikalikan dengan matriks $U$, vektor $v^T_1 \text{ dan } v^T_2$ akan dirotasi ke orientasi minor axis dan major axis elips di $\R^3$
